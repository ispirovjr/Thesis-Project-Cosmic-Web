```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations


fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(projection='3d')
ax.set_aspect("equal")






ang = np.pi/4

u, v = np.mgrid[0:2*np.pi:20j, ang:np.pi-ang:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
ax.plot_wireframe(x, y, z, color="r",cstride=1,rstride=1)


u, v = np.mgrid[0:2*np.pi:10j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
ax.plot_surface(x, y, z, color="b",alpha=0.5)


plt.show()
```

```python
AngS = np.linspace(0,2*np.pi)

AngS2 = np.linspace(-np.pi/2+ang,np.pi/2-ang)

plt.figure(figsize=(10,10))

plt.plot(np.cos(AngS),np.sin(AngS),c='r')

#plt.plot(np.cos(AngS2),np.sin(AngS2))
plt.fill_between(np.append(np.cos(AngS2),0),np.append(np.sin(AngS2),(0)),-np.append(np.sin(AngS2),(0)),color='b')
plt.fill_between(-np.append(np.cos(AngS2),0),np.append(np.sin(AngS2),(0)),-np.append(np.sin(AngS2),(0)),color='b',label="Accepted Values")
#plt.xlabel("RA")
plt.ylabel("dec")
plt.title("Allowed Values")
plt.legend()
plt.show()
```
